Data Collection And PreProcssing

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data= pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

# **Feature slection**

```python
selected_features=[ 'genres','keywords','tagline','cast','director']
```

```python
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
combined_features = movies_data ['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

# **Vectorising the combined string**

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

# **Getting Similarity Using Cosine Similarity**

```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

**Taking User Input**

```python
movie_name= input('Enter Your movie for recommendation: ')
```

```python
list_of_all_titles = movies_data['title'].tolist()
```

```python
close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
print(close_match)
```

```python
top_match = close_match[0]
print(top_match)
```

```python
index_of_the_movie = movies_data[movies_data.title ==top_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#list of recommendation
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
recommendations = sorted(similarity_score,key=lambda x:x[1],reverse=True)
```

```python
print(recommendations)
```

```python
print('Movies recommendations: \n')

i =1
for movie in recommendations:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<11):
   print (i,'.',title_from_index)
   i+=1
```
